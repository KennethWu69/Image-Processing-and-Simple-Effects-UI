```python
import cv2
from matplotlib import pyplot as plt
```

```python
##### resize #####
backimage_path = 'test.jpg'
backimg = cv2.imread(backimage_path)
h, w = backimg.shape[0], backimg.shape[1]
backimg = cv2.resize(backimg, (int(w/10), int(h/10)), interpolation=cv2.INTER_AREA)
print(backimg.shape)
cv2.imshow('output', backimg)
cv2.waitKey(0)
cv2.destroyAllWindows()
# input_batch = transform(img).to(device)
```
